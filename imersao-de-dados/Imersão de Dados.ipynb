## Dia 2 : Tratamento dos dados

```python
#Separando os valores por string e transformando em um DataFrame
dados["Valor"].str.split(expand=True)
```

```python
#Garantindo que todos os dados da coluna 0 são inválidos 
dados["Valor"].str.split(expand=True)[0].unique()
```

```python
#Garantindo que todos os dados da coluna 1 são inválidos
dados["Valor"].str.split(expand=True)[2].unique()
```

```python
dados[["Moeda", "Valor_Anuncio", "Tipo_Pagamento"]] = dados["Valor"].str.split(expand=True)
dados.head()
```

```python
dados[dados["Tipo_Pagamento"].isnull()]["Tipo_Pagamento"].unique()
```

```python
dados_vendas = dados[dados["Tipo_Pagamento"].isnull()]
dados_vendas["Valor_Anuncio"].str.replace(".","").astype(float)
```

```python
dados_vendas["Valor_Anuncio"] = dados_vendas["Valor_Anuncio"].str.replace(".", "").astype(float)
dados_vendas.info()
```

```python
dados_vendas.Valor_Anuncio.plot.hist(bins = 50)
```

```python
#Formatando a mostragem dos dados nas opções do pandas
pd.set_option("display.precision", 2, "display.float_format", lambda x:'%.2f' % x)
dados_vendas.describe(include = "all")
```

```python
dados_vendas
```

```python
sns.set()
plt.figure(figsize=(10,8))
ax = sns.histplot(data = dados_vendas, x = "Valor_Anuncio", kde = True)
ax.set_title("Histograma de valores dos Imóveis em São Paulo")
ax.set()
plt.xlim((-50, 10000000))
plt.show()
```

```python
dados_vendas ["Valor_m2"] = dados_vendas["Valor_Anuncio"] / dados_vendas["Metragem"]
dados_vendas.head()
```

```python
dados_vendas.groupby("Bairro").mean()
```

```python
dados_vendas.groupby("Bairro").median()
```

```python
dados_bairro_soma = dados_vendas.groupby("Bairro").sum()
dados_bairro_soma["Valor_m2_bairro"] = dados_bairro_soma["Valor_Anuncio"] / dados_bairro_soma["Metragem"]
dados_bairro_soma
```

```python
topbairros = dados_vendas["Bairro"].value_counts()[:10].index
```

```python
dados_bairro_soma.reset_index(inplace = True )#(inplace = True)Altera diretamente no DataFrame
```

```python
dados_bairro_soma.query("Bairro in @topbairros")
```

```python
plt.figure(figsize=(10,8))
ax = sns.barplot(x = "Bairro" , y = "Valor_m2_bairro", 
                 data = dados_bairro_soma.query("Bairro in @topbairros"), ci = False)
plt.ylabel("Valor por m²")
ax.tick_params(axis = 'x', rotation = 45)
```
